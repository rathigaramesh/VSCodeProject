```python
import os
import json
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

FOLDER = r'S:\AI_RELATED\AI_PROJECTS\FakeNewsNet\code\fakenewsnet_dataset\politifact\fake'

titles = []

for subfolder in os.listdir(FOLDER):
    content_file = os.path.join(FOLDER, subfolder, 'news content.json')
    if os.path.exists(content_file):
        try:
            with open(content_file, 'r', encoding='utf-8') as f:
                content = json.load(f)
                title = content.get('title', '')
                if title:
                    titles.append(title)
        except Exception as e:
            print(f"Error reading {content_file}: {e}")

# Combine all titles into one text
text_blob = ' '.join(titles)

# Create WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_blob)
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Fake News Titles")
plt.show()
```
